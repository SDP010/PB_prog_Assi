```python
#1. Adding two matrices

r = int(input("Enter the number of rows:"))
c = int(input("Enter the number of columns:"))

#taking input for the first matrix

print("Enter the elements of A matrix row wise : ")
matrix_A = []
for i in range(r):         
    a =[]
    for j in range(c):  
         a.append(int(input()))
    matrix_A.append(a)
print("A = ", matrix_A)

#taking input for the second matrix

print("Enter the elements of B matrix row wise : ")
matrix_B = []
for i in range(r):         
    a =[]
    for j in range(c):  
         a.append(int(input()))
    matrix_B.append(a)
print("B = ", matrix_B)

#Initialising a null matrix of order r x c
result = []
for i in range(r):
    row=[]
    for j in range(c):
        row.append(0)
    result.append(row)
    
#adding the two matrices

for i in range(len(matrix_A)):  
    for j in range(len(matrix_A[0])):
        result[i][j] = matrix_A[i][j] + matrix_B[i][j]
        
#printing the matrices

print("Result is : ")
for i in result:
    print(i)
```

```python
# Python Program to Multiply Two Matrices

m = int(input("Enter the number of rows of first matrix :"))
n= int(input("Enter the number of columns of first matrix and number of rows of second matrix :"))
p = int(input("Enter the number of coloumn of second matrix :"))

#taking input for the first matrix

print("Enter the elements of A matrix row wise : ")
matrix_A = []
for i in range(m):         
    a =[]
    for j in range(n):  
         a.append(int(input()))
    matrix_A.append(a)
print("A = ", matrix_A)

#taking input for the second matrix

print("Enter the elements of B matrix row wise : ")
matrix_B = []
for i in range(n):         
    a =[]
    for j in range(p):  
         a.append(int(input()))
    matrix_B.append(a)
print("B = ", matrix_B)

#Initialising a null matrix of order r x c
result = []
for i in range(m):
    row=[]
    for j in range(p):
        row.append(0)
    result.append(row)
    
#multiplying the two matrices

for i in range(len(matrix_A)):  
    
    for j in range(len(matrix_B[0])):
        
        for k in range(len(matrix_B)) :
            result[i][j] += matrix_A[i][k] * matrix_B[k][j]
        
#printing the matrices

print("Result is : ")
for i in result:
    print(i)
```

```python
#4. program to sortwords in algebraic order

def sort_words(n) :
    l=[]
    print("Enter the words : ")
    for i in range(n):
        l.append(input())
    print("Given words are : ", l)
    print("After sorting, words are : ", l , end ="")
    return (l.sort())

try :
    length = int(input('enter length of list of strings: '))
    sort_words(length)
except Exception as e : 
    print(e)

```

```python
#5. Python Program to Remove Punctuation From a String?

def remove_punc(s):
    without_punc =''
    punc='~`!@$#%^&*_-+=?><.,:;""'
    for i in s :
        if not i in punc :
            without_punc +=  i
    return without_punc
try :
    st = input("Enter the string : ")
    print(remove_punc(st))
except Exception as e :
    print(e)
```
